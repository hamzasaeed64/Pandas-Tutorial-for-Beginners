```python
## Converting the Whole Size Column into Bytes

# Lets Define a Function to Convert the Size Column into Bytes

def size_to_bytes(size):
    if isinstance(size, str):
        if 'k' in size:
            return float(size.replace('k', '')) * 1024
        elif 'M' in size:
            return float(size.replace('M', '')) * 1024 * 1024
        elif 'Varies with device' in size:
            return np.nan
    return size
```

```python
# Before
df['Size']
```

```python
# After Applying this Function to the Size Column
df['Size'] = df['Size'].apply(size_to_bytes)
df['Size']
```

```python
df.head()
```

```python
# We See that Column Values are Converted but The Column Name is Still Size so we Chnage it by Size in Binary
df.rename(columns={'Size': 'Size_in_Bytes'}, inplace=True)
```

```python
# Now Lets Check the Column Name
df.columns
```

```python
df['Size_in_Bytes'] = pd.to_numeric(df['Size_in_Bytes'], errors='coerce')
df['Size_in_Mb'] = df['Size_in_Bytes'].apply(lambda x: x/(1024 * 1024))
df['Size_in_Kb'] = df['Size_in_Bytes'].apply(lambda x: x/(1024))
```

```python
## Removing the '+' and ',' from the Installs Column
df['Installs'] = df['Installs'].apply(lambda x: x.replace('+', '') if '+' in str(x) else x)
df['Installs'] = df['Installs'].apply(lambda x: x.replace(',', '') if ',' in str(x) else x)

```

```python
df['Installs'].unique()
```

```python
# Replace non-numeric values with 0 and convert to integers
df['Installs'] = df['Installs'].apply(lambda x: x if x.isdigit() else '0').astype(int)
```

```python
df['Installs'].value_counts()
```

```python
# Dealing with the Price Column
df['Price'].value_counts()
```

```python
# if we see the Price Column we see that there is a $ sign in the Column so we need to remove it
df['Price'] = df['Price'].apply(lambda x: x.replace('$', '') if '$' in str(x) else x)
df['Price'] = pd.to_numeric(df['Price'], errors='coerce')
```

```python
# Since the $ sign has already been removed, we cannot check for it directly.
# Instead, we can count the number of non-zero values in the 'Price' column.
non_zero_prices = df['Price'].loc[df['Price'] > 0].count()
print(f"Number of values that originally had $ sign: {non_zero_prices}")
```

```python
# Count the Number of Values in 'Price' Columns that are 0 but did not have a $ sign
zero_prices = df['Price'].loc[df['Price'] == 0].count()
print(f"Number of values that did not have $ sign: {zero_prices}")
print(f"Total number of values: {len(df)}")
```

```python
# we can calculate the pearson correlation coefficient using scipy as well as follows

# this is to install scipy if you have not done it before
# pip install scipy 
from scipy import stats

# remove rows containing NaN or infinite values (Important to calculate Pearson's R)
df_clean = df.dropna(subset=['Reviews', 'Installs'])

# ensure the columns are numeric
df_clean['Reviews'] = pd.to_numeric(df_clean['Reviews'], errors='coerce')
df_clean['Installs'] = pd.to_numeric(df_clean['Installs'], errors='coerce')

# calculate Pearson's R between Reviews and Installs
pearson_r, _ = stats.pearsonr(df_clean['Reviews'], df_clean['Installs'])
print(f"Pearson's R between Reviews and Installs: {pearson_r:.4f}")
```

`2.3. Duplicates`
Removing duplicates is one of the most important part of the data wrangling process, we must remove the duplicates in order to get the correct insights from the data.
If you do not remove duplicates from a dataset, it can lead to incorrect insights and analysis.
Duplicates can skew statistical measures such as mean, median, and standard deviation, and can also lead to over-representation of certain data points.
It is important to remove duplicates to ensure the accuracy and reliability of your data analysis.
